```python
lines = []
# Read and clean the lines from the file
with open("input.txt") as f:
    lines = [line.strip() for line in f if line.strip()]

sum = 0

# Iterate through each line and character to find patterns
for j in range(len(lines)):
    line = lines[j]
    for i in range(len(line)):
        # Check for horizontal patterns
        if i + 3 < len(line):
            if line[i : i + 4] in ["XMAS", "SAMX"]:
                sum += 1
        
        # Check for diagonal-right patterns
        if i + 3 < len(line) and j + 3 < len(lines):
            diagonal_right = (
                lines[j][i]
                + lines[j + 1][i + 1]
                + lines[j + 2][i + 2]
                + lines[j + 3][i + 3]
            )
            if diagonal_right in ["XMAS", "SAMX"]:
                sum += 1
        
        # Check for vertical patterns
        if j + 3 < len(lines):
            down = (
                lines[j][i]
                + lines[j + 1][i]
                + lines[j + 2][i]
                + lines[j + 3][i]
            )
            if down in ["XMAS", "SAMX"]:
                sum += 1
        
        # Check for diagonal-left patterns
        if i - 3 >= 0 and j + 3 < len(lines):
            diagonal_left = (
                lines[j][i]
                + lines[j + 1][i - 1]
                + lines[j + 2][i - 2]
                + lines[j + 3][i - 3]
            )
            if diagonal_left in ["XMAS", "SAMX"]:
                sum += 1

print(sum)

#Part 2

lines = []
with open("input.txt") as f:
    lines = f.readlines()

sum = 0

for j in range(1, len(lines) - 1):
    line = lines[j]
    for i in range(1, len(line) - 1):
        if line[i] == "A":
            diagonal_left = (lines[j - 1][i - 1], lines[j + 1][i + 1])
            diagonal_right = (lines[j - 1][i + 1], lines[j + 1][i - 1])
            if (diagonal_right == ("M", "S") or diagonal_right == ("S", "M")) and (
                diagonal_left == ("M", "S") or diagonal_left == ("S", "M")
            ):
                sum += 1
print(sum)
```
